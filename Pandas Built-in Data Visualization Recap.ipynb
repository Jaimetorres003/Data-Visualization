# Pandas Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df3 = pd.read_csv('df3')
df3.info()
```

```python
df3.head()
```

```python
# creating a scatter plot of b vs a. I'm going to increase the size of the points and the figure size
df3.plot.scatter(x='a', y='b', c='purple', s=50, figsize=(12,3))
```

```python
# I'm creating a histogram of the 'a' column
df3['a'].plot.hist(bins=12, alpha=1.0)
```

```python
#Another way to create a histogram is to use ggplot
plt.style.use('ggplot')
df3['a'].plot.hist(alpha=0.5, bins=25)
```

```python
#creating a boxplot comparing the a and b columns
df3[['a','b']].plot.box()
```

```python
#creating a kde plot of the 'd' column
df3['d'].plot.kde()
```

```python
#changing the linewidth and making the linestyle dashed
df3['d'].plot.density(lw=5,ls='--')
```

```python
#Creating an area plot of all the columns for just the rows up to 30
df3.iloc[0:30].plot.area(alpha=0.4)
```

```python
#I don't like that the legend overlaps my data. I want to move it over 
f=plt.figure()
df3.iloc[0:30].plot.area(alpha=0.4, ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
# Alternatively can use best
f=plt.figure()
df3.iloc[0:30].plot.area(alpha=0.4, ax=f.gca())
plt.legend(loc='best', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
