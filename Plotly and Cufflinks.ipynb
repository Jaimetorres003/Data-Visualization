```python

from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot
```

```python
#for connecting javascript to notebook
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df=pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2=pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2
```

```python
df.plot()
```

```python
#how to use cufflinks in pyplot
df.iplot()
```

```python
#make a scatterplot, if you don't specify the mode cufflinks will try to link all the points together
df.iplot(kind='scatter', x = 'A', y='B', mode='markers')
```

```python
#bar plot using DF2
df2.iplot(kind='bar', x ='Category', y='Values')
```

```python
#group by or aggregate function
df.sum().iplot(kind='bar')
```

```python
#boxplots
df.iplot(kind='box')
```

```python
df3=pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[500,400,300,200,100]})
```

```python
df3
```

```python
#3D surface plot
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
#histogram for only a single column, A
df['A'].iplot(kind='hist',bins=50)
```

```python
#histogram passing in all columns
df.iplot(kind='hist',bins=50)
```

```python
#spread data used for stocks in finance
df[['A','B']].iplot(kind='spread')
```

```python
#bubbleplot similar to scatter but changes size based on another variable common plots for United Nations reports
df.iplot(kind='bubble', x='A',y='B',size='C')
```

```python
#use with care on large data sets or you may crash your kernel
df.scatter_matrix()
```
